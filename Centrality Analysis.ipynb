```python
%matplotlib inline

import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from operator import itemgetter, attrgetter
```

```python
data= pd.read_csv("D:/College/5th sem/GT project/CSV/CAVIAR_FULL.csv")
```

```python
data
```

```python
data=data.set_index(data.columns[0])
```

```python
data
```

```python
graph=nx.from_numpy_matrix(data.values)
```

```python
graph=nx.relabel_nodes(graph,dict(enumerate(data.columns)))
graph.nodes()
```

```python
pos=dict(zip(list(enumerate(data.columns)),data.values))
#nx.draw_networkx_labels(graph,pos,font_size=20, font_family='sans-serif')
nx.draw_spring(graph,with_labels=True)
plt.figure(figsize=(700,10.0))
plt.show()
```

```python
deg=nx.degree_centrality(graph)
close=nx.closeness_centrality(graph)
bet=nx.betweenness_centrality(graph)
```

```python
sort_deg=sorted(deg.items(),key=itemgetter(1),reverse=True)
sort_close=sorted(close.items(),key=itemgetter(1),reverse=True)
sort_bet=sorted(bet.items(),key=itemgetter(1),reverse=True)
```
